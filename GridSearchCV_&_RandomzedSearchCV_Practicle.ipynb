```
import pandas as pd
```

```
dataset = pd.read_csv("/content/regression2.csv")
```

```
dataset.head(3)
```

```
x = dataset.iloc[:,:-1]
y = dataset["Salary"]
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```
from sklearn.tree import DecisionTreeRegressor
```

```
dt = DecisionTreeRegressor(criterion="squared_error",splitter="best",max_depth=10)
dt.fit(x_train,y_train)
```

```
dt.score(x_train,y_train)*100, dt.score(x_test,y_test)*100
```

```
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```
dt_params = {"criterion":["squared_error", "friedman_mse", "absolute_error", "poisson"],
           "splitter":["best", "random"],
           "max_depth":[i for i in range(2,20)]}
```

```
4 * 2 * 18
```

```
rd = RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=dt_params,n_iter=20)
rd.fit(x_train,y_train)
```

```
rd.best_params_
```

```
rd.best_score_
```

```
gd = GridSearchCV(DecisionTreeRegressor(),param_grid=dt_params)
gd.fit(x_train,y_train)
```

```
gd.best_params_
```

```
gd.best_score_
```
