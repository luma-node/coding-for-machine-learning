```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
dataset = pd.read_csv("/content/IRIS.csv")
```

```
import scipy.cluster.hierarchy as sc
```

```
plt.figure(figsize=(20,20))
sc.dendrogram(sc.linkage(dataset[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']],method='ward',metric='euclidean'))
plt.show()
```

```
from sklearn.cluster import AgglomerativeClustering
```

```
ac = AgglomerativeClustering(n_clusters=2,linkage='single')
dataset["Predict"] = ac.fit_predict(dataset.drop('species', axis=1))
```

```
dataset.head(2)
```

```
sns.pairplot(data=dataset,hue="Predict")
plt.show()
```
